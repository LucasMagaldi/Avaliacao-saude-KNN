# PRE-PROCESSING

SEPARATION VARIABLES IN PREDICTOR AND TARGET

```python
x = df.drop('class', axis = 1)
y = df['class']
```

TRAIN AND TEST

```python
xTrain, xTest, yTrain, yTest = train_test_split(x,y,test_size = 0.2, random_state=7)
```

# PREDICT MACHINE

BUILD A MODEL AND TRAIN DATA

```python
predict =KNeighborsClassifier(n_neighbors=3)
knn = predict.fit(xTrain,yTrain)
knn
```

```python
predictY = knn.predict(xTest)
```

COMPARE ACTUAL AND PREDICT VALUES

```python
actualVSpredict = pd.DataFrame({'Actual: ':yTest, 'Prediction:': predictY})
actualVSpredict.head(20)
```

# AVALIATION

ACCURACY


```python
accuracy_score(yTest,predictY)
```

CLASSIFICATION METRICS

```python
print(classification_report(yTest,predictY))
```
